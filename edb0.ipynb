# Recursividade normal e de cauda

## Soma


```python
def soma_rec(n):
  if n == 0:
    return n;
  return n + soma_rec(n-1)
```

```python
soma_rec(6)
```

```python
def soma_rec_t(n, a = 0):
  if n == 0:
    return a
  return soma_rec_t(n-1, a+n)
```

```python
soma_rec_t(6)
```

## Fatorial

```python
def fat_rec(n):
  if n == 1:
    return n;
  return n * fat_rec(n-1)
```

```python
fat_rec(5)
```

```python
def fat_rec_t(n, a = 1):
  if n == 1:
    return a
  return fat_rec_t(n-1, n*a)
```

```python
fat_rec_t(5)
```

## Multiplicação

```python
def mult_rec(n1, n2):
  if n2 == 1:
    return n1
  elif n1 == 0 or n2 == 0:
    return 0
  return n1 + mult_rec(n1, n2-1)
```

```python
mult_rec(4, 43)
```

```python
def multi_rec_t(n1, n2):
  def mult_rec_t2(n1,n2,a=0):
    if n2 == 0:
      return a
    return mult_rec_t2(n1, n2-1, n1+a)
  if n1 == 0 or n2 == 0:
    return 0
  return mult_rec_t2(n1,n2)
```

```python
multi_rec_t(35,34)
```
